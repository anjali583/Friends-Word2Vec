```python
import pandas as pd
import numpy as np
import gensim
import os
```

```python
import nltk
nltk.download('punkt')
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
story=[]
f=open('/content/sample_data/data/Friends_Transcript.txt', encoding='latin-1')
corpus=f.read()
raw_sent=sent_tokenize(corpus)
for sent in raw_sent:
    story.append(simple_preprocess(sent))
```

```python
story
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2
)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model.wv.most_similar('kill')
```

```python
model.wv.doesnt_match(['umbrella','home','breathe'])
```
